## Teste de hipótese

**H<sup>null</sup>**
>**A distribuição da taxa de gorjeta é a mesma nos dois grupos**

**H<sup>alt</sup>**
>**A distribuição da taxa de gorjeta não é a mesma nos dois grupos**

```python
from scipy.stats import ranksums
```

```python
com_sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```python
sem_sobremesa = gorjetas.query("sobremesa == 'Nao'").porcentagem
```

```python
sem_sobremesa
```

```python
r = ranksums(com_sobremesa, sem_sobremesa) 
```

```python
print(f"O valor do p-value é {r.pvalue}.\n Maio valor para alternativa ser aceite para a hipótese alternativa é de no máximo: 0.05")
```

**H<sup>null</sup>**
>**A distribuição da taxa de gorjeta é a mesma nos dois grupos**

## Teste de hipótese

**H<sup>null</sup>**
>**A distribuição do valor da conta é igual no sábado e no domingo**

**H<sup>alt</sup>**
>**A distribuição do valor da conta não é igual no sábado e no domingo**

```python
valor_conta_domingo = gorjetas.query("dia_da_semana == 'Domingo'").valor_da_conta
```

```python
valor_conta_sabado = gorjetas.query("dia_da_semana == 'Sábado'").valor_da_conta
```

```python
r2 = ranksums(valor_conta_domingo, valor_conta_sabado)
print(f'O valor do p-value é {r2.pvalue}')
```

**H<sup>alt</sup>**
>**A distribuição do valor da conta não é igual no sábado e no domingo**
